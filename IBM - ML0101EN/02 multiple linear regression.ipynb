```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
```

```python
# download the data
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv')
df.head()
```

```python
# now consider just the columns of interest
cdf = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_CITY', 'FUELCONSUMPTION_HWY', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS']]
cdf.head(9)
```

```python
# plot engine size vs emission values
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
# split data set into train (80%) and test (20%) sets
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
# train data distribution
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
# use sklearn package to model data
from sklearn import linear_model
regr = linear_model.LinearRegression()
parameters_1 = ['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']
x = np.asanyarray(train[parameters_1])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (x, y)
# Output the coefficients i.e. the parameters of the fit line
print ('Coefficients: ', regr.coef_)
```

```python
# prediction using FUELCONSUMPTION_COMB
y_hat= regr.predict(test[parameters_1])
x = np.asanyarray(test[parameters_1])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance regression score (note that 1 is perfect prediction)
print('Variance score: %.2f' % regr.score(x, y))
```

```python
# Repeat using different parameters
regr = linear_model.LinearRegression()
parameters_2 = ['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_CITY','FUELCONSUMPTION_HWY']
x = np.asanyarray(train[parameters_2])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (x, y)
# Output the coefficients i.e. the parameters of the fit line
print ('Coefficients: ', regr.coef_)
```

```python
# prediction using FUELCONSUMPTION_CITY and FUELCONSUMPTION_HWY
y_hat= regr.predict(test[parameters_2])
x = np.asanyarray(test[parameters_2])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance regression score (note that 1 is perfect prediction)
print('Variance score: %.2f' % regr.score(x, y))
```
