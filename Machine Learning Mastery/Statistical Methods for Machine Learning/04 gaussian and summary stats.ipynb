# Gaussian and Summary Stats

```python
# generate and plot an idealised gaussian (aka normal) distribution
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# define the axes
x_axis = arange(-3, 3, 0.001)
y_axis = norm.pdf(x_axis, 0, 1)

# plot the data 
pyplot.plot(x_axis, y_axis)
pyplot.show()
```

```python
# the randn function generates a specified number (e.g. 10000_ of random numbers from a gaussian distribution with a mean of 0 and a standard deviation of 1
# these numbers can be scaled to a gaussian of our choosing by rescaling the numbers
from numpy.random import randn
data = randn(10000)

# this can be made consistent by adding the desired mean (e.g. 50) and multiplying the value by the standard deviation (e.g. 5)
data = 5 * randn(10000) + 50
```

```python
# generate a set of random gaussians and plot the sample
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot

# seed the random number generator
seed(1)

# generate univariate observations
data = 5 * randn(10000) + 50

# histogram of generated data
pyplot.hist(data)
pyplot.show()
```

```python
# note that, in the example above, the plotting function chooses an arbitrary sized bucket for splitting up the data
# update the example with a more refined plot
pyplot.hist(data, bins=100)
pyplot.show()
```

```python
# in a gaussian distribution, the central tendency is called the mean (more formally: arithmetic mean) and is one of hte two main parameters that defines any gaussian distribution
from numpy import mean
result = mean(data)
print('Mean: %.3f' % result)
```

```python
# in the case of outliers or a non-gaussian distribute, an alternate and commonly used central tendency is the median.  This is calculated by sorting all the data and locating the middle value (if there is an even number of observations, calculate the mean of the middle two)
from numpy import median
result = median(data)
print('Median: %.3f' % result)
```

```python
# the variance describes how much on average observations vary from the mean value i.e. the measure of spread of a distrubution
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# x-axis for the plot
x_axis = arange(-3, 3, 0.001)

# first plot a gaussian with a low variance (blue)
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.5))

# now plot a gaussian with a high variance (orange)
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 1))
pyplot.show()
```

```python
# the variance of a data sample drawn from a gaussian distribution is calculated as the average squared difference of each observation in the sample from the sample mean
from numpy.random import seed
from numpy.random import randn
from numpy import var

# seed the random number generator
seed(1)

# generate univariate observations
data = 5 * randn(10000) + 50

# calculate variance
result = var(data)
print('Variance: %.3f' % result)
```

```python
# the standard deviation is the square root of the variance, and is (along with the mean) one of the two key parameters used to specify any gaussian distribution
from numpy import std
result = std(data)
print('Standard Deviation: %.3f' % result)
```

# EXTENSION

```python
def data_summary(data, defined_mean, defined_std):
    import numpy as np
    from matplotlib import pyplot

    # plot graph
    pyplot.hist(data, bins='auto')
    pyplot.show()

    # display statistics
    print('\nDefined Mean:       %.3f' % defined_mean)
    print('Defined Std:        %.3f' % defined_std)
    print('\nMean:               %.3f' % np.mean(data))
    print('Median:             %.3f' % np.median(data))
    print('Variance:           %.3f' % np.var(data))
    print('Standard Deviation: %.3f' % np.std(data))
```

```python
import numpy as np

# generate observations without seeding the random number generator
defined_mean = np.random.random() * 100
defined_std = np.random.random() * 10
data = defined_std * np.random.randn(10000) + defined_mean

# display data summary
data_summary(data, defined_mean, defined_std)
```
