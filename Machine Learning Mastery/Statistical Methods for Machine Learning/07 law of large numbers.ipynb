# Law of Large Numbers

As the size of the sample increases, the mean value of the sample will better approximate the mean or expected value in the population.  As the sample size approaches infinity, the sample mean will converge to the populations mean.

The observations in the sample must be independent.

Note that when working with extremely large samples of observations, we also increase the likelihood of seeing something unusual.

We must be wary not to make inferences from single cases.

```python
# idealized population distribution
from numpy import arange
from matplotlib import pyplot as plt
from scipy.stats import norm

# x-axis for the plot
xaxis = arange(30, 70, 1)

# y-axis for the plot
yaxis = norm.pdf(xaxis, 50, 5)

# plot ideal population
plt.plot(xaxis, yaxis)
plt.show()
```

Now forget what we know about the population and make independent random samples from the population.  Create samples of different sizes and calculate the mean.

```python
# demonstrate the law of large numbers
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import array
from matplotlib import pyplot as plt

# seed the random number generator
seed(1)

# sample sizes
sample_sizes = []
for x in range (10, 20000, 200):
    sample_sizes.append(x)

# generate samples of different sizes and calculate their means
# the NumPy randn function returns a list of the specified length of normally distributed values with mean 0 and standard deviation 1.0
# we scale them to have a mean of 50 and standard deviation of 5
means = [mean(5 * randn(sample_size) + 50) for sample_size in sample_sizes]

# plot sample mean error vs sample size
# (we know the true mean is 50 so subtract 50 to determine error)
plt.scatter(sample_sizes, array(means)-50)
plt.show()

# generally, larger samples sizes have less error
```

Implications in Machine Learning

- Training data - the data used to train the model must be representative of the observations from the domain.  Keep this in mind during data collection, data cleaning and data preparation
- Test data - this is often neglected with the blind use of a 80/20 split for train/test data, or the blind use of a 10-fold cross-validation, even on datasets where the size of 1/10th of the available data may not be a suitable representative of observations from the problem domain
- Model Skill Evaluation - consider the law of large numbers when presenting the estimated skill of a model on unseen data. It highlights the need to develop a sample of multiple independent (or close to) evaluations of a given model such that the mean reported skill from the sample is an accurate enough estimate of population mean
